```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load the breast cancer dataset
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a logistic regression model
model = LogisticRegression(max_iter=10000)

# Initialize the list of features
remaining_features = list(X_train.columns)

# Initialize a list to store accuracies
accuracies = []

# Loop through the feature elimination process
while len(remaining_features) > 1:
    # Fit the model on the current set of features
    model.fit(X_train[remaining_features], y_train)
    
    # Get the coefficients of the current model
    coefs = pd.Series(model.coef_[0], index=remaining_features)
    
    # Print the coefficients of the current model
    print(f"\nCurrent Features: {remaining_features}")
    print("Coefficients:\n", coefs)
    
    # Identify the least important feature (smallest absolute coefficient)
    least_important_feature = coefs.abs().idxmin()
    print(f"Least important feature to remove: {least_important_feature}")
    
    # Remove the least important feature from the list of remaining features
    remaining_features.remove(least_important_feature)
    
    # Refit the model with the updated set of features
    model.fit(X_train[remaining_features], y_train)
    
    # Make predictions using the updated model and features
    y_pred = model.predict(X_test[remaining_features])
    
    # Calculate accuracy with the current set of features
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)
    print(f"Accuracy after removing '{least_important_feature}': {accuracy:.4f}")
    
    # Optional: visualize the current coefficients
    coefs.plot(kind='bar', title="Feature Coefficients", color='skyblue')
    plt.tight_layout()
    plt.show()

# Final model after elimination
print("\nFinal selected features:", remaining_features)

# Final accuracy
model.fit(X_train[remaining_features], y_train)
y_pred = model.predict(X_test[remaining_features])
final_accuracy = accuracy_score(y_test, y_pred)
print(f"\nFinal model accuracy: {final_accuracy:.4f}")
# Plot accuracy vs number of features
plt.figure(figsize=(10, 6))
plt.plot(range(len(X.columns), 1, -1), accuracies)
plt.xlabel('Number of Features')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Number of Features')
plt.grid(True)
plt.show()
```
